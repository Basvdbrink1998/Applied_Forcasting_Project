```r
library(fpp3)
library(tsintermittent)
```

```r
# import all available data 
sales_validation <- read.csv('sales_train_validation_afcs2021.csv')
sample_submission <- read.csv('sample_submission_afcs2021.csv')
```

```r
sales_df <- as.data.frame( t(sales_validation))

fc_tbats = sample_submission
fc_holt = sample_submission


for (i in 1:823) {
  # R dataframe index is in format Vx with x being the id. Using this code, the correct row can be selected
  in_id = paste('V',i, collapse='')
  in_id = gsub(" ", "", in_id)

  # create time series 
  sales_ts = ts(sales_df[in_id], frequency=365) 
  
  # fit and forecast
  tbats = tbats(sales_ts,use.box.cox=FALSE, use.trend=FALSE, use.damped.trend = FALSE)
  fc_tbats[i,2:29] = forecast(tbats)$mean
  fc_holt[i,2:29] = holt(sales_ts, h=28)$mean
}

```

```r
submission_holt = sample_submission
submission_tbats = sample_submission
for (i in 1:823) {
  submission_holt[i,2:29] = fc_holt[i,2:29]
  submission_tbats[i,2:29] = fc_tbats[i,2:29]
}

write.csv(submission_holt, 'forecast_holt.csv', row.names=FALSE)
write.csv(submission_tbats, 'forecast_tbats.csv', row.names=FALSE)
```
