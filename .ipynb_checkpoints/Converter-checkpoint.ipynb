```python
calendar = pd.read_csv("calendar_afcs2021.csv")
sell_prices = pd.read_csv("sell_prices_afcs2021.csv", index_col=0)

sample_submission = pd.read_csv("sample_submission_afcs2021.csv", index_col=0)

train_data = pd.read_csv("sales_train_validation_afcs2021.csv", index_col=0)
test_data = pd.read_csv("sales_test_validation_afcs2021.csv", index_col=0)
```

```python
calendar['date'] = pd.to_datetime(calendar['date'], infer_datetime_format=True)
```

```python
zip_iterator = zip(calendar['d'],calendar["date"])
date_converter = dict(zip_iterator)
```

```python
try:
    train_data = train_data.rename(columns = date_converter)
except:
    pass
try:
    test_data = test_data.rename(columns = date_converter)
except:
    pass
```

```python
indx = 110
x = train_data.iloc[indx]
x.plot()
```

```python
# SES example
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
# contrived dataset
data = x
data
# fit model
model = SimpleExpSmoothing(data)
model_fit = model.fit()
# make prediction
yhat = model_fit.predict(len(data), len(data) + 27)
yhat
```

```python
# SARIMA example
from statsmodels.tsa.statespace.sarimax import SARIMAX
from random import random
# contrived dataset
data = x
# fit model
model = SARIMAX(data, order=(1, 1, 1), seasonal_order=(0, 0, 0, 0))
model_fit = model.fit(disp=False)
# make prediction
yhat = model_fit.predict(len(data), len(data) + 27)
print(yhat)
```

```python
rms = mean_squared_error(test_data.iloc[indx], yhat, squared=False)
rms
```
